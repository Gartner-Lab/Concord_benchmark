```r
library(Seurat)
library(monocle3)
pyenv = 'cellpath'

library(anndata)
library(reticulate)
use_condaenv(pyenv, required = TRUE)
```

```r
eset = readRDS('../data/intestine_dev/R_obj/eset.rds')
seu_int = readRDS('../data/intestine_dev/R_obj/gut_integrated.rds')
clist = readRDS('../data/intestine_dev/R_obj/clist.rds')
```

```r
print(dim(eset))
print(dim(seu_int))
```

```r
library(VisCello)
```

```r
fmeta = eset@featureData@data
pmeta = eset@phenoData@data

adata <- AnnData(
    X =  t(eset@assayData$exprs),
    obs = pmeta,
    var = fmeta,
)
# add obsm
for (key in names(seurat_obj@reductions)) {
    adata$obsm[[key]] <- seurat_obj@reductions[[key]]@cell.embeddings
}
```

```r
seurat_integrated_pca = clist[["All cells (batch corrected)"]]@proj[['PCA']]
adata$obsm[["Seurat_int_PCA"]] = seurat_integrated_pca
```

```r
adata$write_h5ad('../data/intestine_dev/adata_Huycke_1119.h5ad')
```
