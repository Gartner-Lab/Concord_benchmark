```r

library(VisCello)
pyenv = 'cellpath'

library(anndata)
library(reticulate)
use_condaenv(pyenv, required = TRUE)
```

```r
eset = readRDS('../data/McMahon_kidney/GSE129798/kidney_cello/eset.rds')
clist = readRDS('../data/McMahon_kidney/GSE129798/kidney_cello/clist.rds')
```

```r
fmeta = eset@featureData@data
pmeta = eset@phenoData@data

adata <- AnnData(
    X =  t(eset@assayData$exprs),
    obs = pmeta,
    var = fmeta,
)

# Initialize obsm attribute
adata$obsm <- list()

# Save proj from clist[['All cells']]@proj to adata.obsm
for (proj in names(clist[['All cells']]@proj)) {
    adata$obsm[[proj]] = clist[['All cells']]@proj[[proj]]
}
```

```r
adata$write_h5ad('../data/McMahon_kidney/GSE129798/Ransick_adata.h5ad')
```
